```python
import pandas as pd
import psycopg2
import os
import matplotlib as plt
import pyarrow.parquet as pq
import time
```

```python
#df.head()
#df['Coluna T1'] = 'A'
start = time.time()
df.loc[(df['trip_distance'] > 1.00) & (df['total_amount'] > 15.00), 'Coluna T1'] = 'Viagens baratas e longas'
df.loc[(df['trip_distance'] <= 1.00) & (df['total_amount'] <= 15.00), 'Coluna T1'] = 'Viagens caras e curtas'
end = time.time()
print(end - start)
```
